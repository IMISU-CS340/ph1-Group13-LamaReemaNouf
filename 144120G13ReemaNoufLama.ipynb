### part one A*:


#####  Heuristic function :
It’s a function that estimates the actual cost to decide which states leads to goal.
There are many ways to calculate the heuristic function.



##### A* : 
A searching algorithm gives the best path in graph by using a function of heuristic h(n) + actual cost g(n) to avoid the expanding of expensive path.  


Let’s apply A* algorithm in maze problem:
 
 
The maze is made of 4x4 squares; we have three types of squares:


1-	Green square represents legal move.


2-	Red square represents a wall (can’t move throw it).


3-	Gray represents fixed place of start and end.



And you have 4 actions: left, right, up, down.



Our goal is to reach the end with minimum number of moves.

```python
class NodeA:
    def __init__(self,x,y,notWall):
        self.x = x
        self.y = y
        self.notWall = notWall
        self.parent = None
        self.g = 0
        self.h = 0
        self.f = self.g+self.h #evalution function
        self.adjecnts = []

    def printCoordanites(self):
        print( '(',self.x,',',self.y,')' )
        
```

we will consider each square of maze as node.

```python
class PriorityQueue: 
            
    def __init__(self): 
        self.queue = [] 
   
    def inqueue(self, Node): 
        self.queue.append(Node) 
  

    #return node with minimum evlution function
    def dequeue(self): 
            min = 0
            for i in range(len(self.queue)): 
                if (self.queue[i].f < self.queue[min].f): 
                    min = i 
            node = self.queue[min]
            del self.queue[min] 
            return node
```

we need priority queue for (( openList )) that we will create it later.

**" inqueue " method:** insert new node to the end of the list.


**" dequeue " method:** search for the node with minimum evalution function (f) value then take it out form the list.

```python
import copy

class Maze:
    def __init__(self): 
        self.nodes = [] 
        
        #you can change the hight and width of the maze 
        self.maze_height = 4
        self.maze_width = 4
        self.start = None
        self.end = None
    
    def initMaze(self):
            
        # try to change walls place !!!
        walls = [(1,1),(2,2),(4,3),(2,4)]
        
        
        #Assume our maze starts from 1
        for x in range(1,self.maze_width+1):
            for y in range(1,self.maze_height+1):
                if ((x,y) in walls):
                    notWall = False
                else:
                    notWall = True
               
                self.nodes.append(NodeA(x,y,notWall))
                
        #Note:you have to change the start and end position if you changed the maze's hight and width 
        self.start = self.nodes[self.getindex(1,4)]
        self.end = self.nodes[self.getindex(4,1)]
        self.start.h = self.calcHeuristic(self.start)
        
    def calcHeuristic(self,node):
      
        #there are other ways to calculate the heuristic function(EX:stright line distance)
        return (abs(node.x - self.end.x) + abs(node.y - self.end.y))
    
    def calcActual(self,node):
        return (node.g)+1
    
    def getAdjecnt(self,node):
            
        # if the node in range and not wall, calculate g,h and assigned the parent of this node
        if ( node.x < 4 and self.nodes[self.getindex((node.x)+1,node.y)].notWall):
                node.adjecnts.append(self.nodes[self.getindex((node.x)+1,node.y)])
                self.nodes[self.getindex((node.x)+1,node.y)].g = self.calcActual(node)
                self.nodes[self.getindex((node.x)+1,node.y)].h = self.calcHeuristic(node)
                self.nodes[self.getindex((node.x)+1,node.y)].parent = copy.copy(node)    
                 
        if (node.y < 4 and self.nodes[self.getindex(node.x,(node.y)+1)].notWall):
                node.adjecnts.append(self.nodes[self.getindex(node.x,(node.y)+1)])
                self.nodes[self.getindex(node.x,(node.y)+1)].g = self.calcActual(node)
                self.nodes[self.getindex(node.x,(node.y)+1)].h = self.calcHeuristic(node)
                self.nodes[self.getindex(node.x,(node.y)+1)].parent = copy.copy(node)
                    
        if (node.x > 1 and self.nodes[self.getindex((node.x)-1,node.y)].notWall):
                node.adjecnts.append(self.nodes[self.getindex((node.x)-1,node.y)]) 
                self.nodes[self.getindex((node.x)-1,node.y)].g = self.calcActual(node)
                self.nodes[self.getindex((node.x)-1,node.y)].h = self.calcHeuristic(node)
                self.nodes[self.getindex((node.x)-1,node.y)].parent = copy.copy(node)
                    
        if (node.y > 1 and self.nodes[self.getindex(node.x,(node.y)-1)].notWall):
                node.adjecnts.append(self.nodes[self.getindex(node.x,(node.y)-1)])
                self.nodes[self.getindex(node.x,(node.y)-1)].g = self.calcActual(node)
                self.nodes[self.getindex(node.x,(node.y)-1)].h = self.calcHeuristic(node)
                self.nodes[self.getindex(node.x,(node.y)-1)].parent = copy.copy(node)

       
    def getindex(self,x,y):
        for i in range(len(self.nodes)):
            if(( self.nodes[i].x == x )and (self.nodes[i].y == y)):
           
              return i
            
        
    def path(self,node):
        if ((node.x == self.start.x) and( node.y == self.start.y)):
            
            return
        else:
            self.path(node.parent)
            node.printCoordanites()
```

when all nodes are created and inserted we define the start and end position, and calculate the heuristic value of start node.

**" calcHeuristic " method:** calculate heuristic value based on Manhattan distance that allows you to move from a node to the end node without restrictions (move throw walls) using the above formela.


**" calcActual " method:** clculate the acttual cost from start node to given node by adding 1 per move.


**" getAdjecnt " method:** fill the (( adjecnts )) list with the successor of each node by checking if the seccssor in range of maze and not wall.


**" getindex " method:** search in (( nodes )) list for a match of the x and y coordinates and return the index of this node.

```python
#these two list are main componant in A*
openList = PriorityQueue()
closedList = []


maze = Maze()
maze.initMaze()

openList.inqueue(maze.start)

#  A* algorithm is start !!!
while openList:
    
    square = openList.dequeue()
   
    if (square is maze.end):
        maze.path(square)
        break
        
    elif(square in closedList):
        continue
    
    maze.getAdjecnt(square)   
    for i in range(len(square.adjecnts)):
        openList.inqueue(square.adjecnts[i])
        
    closedList.append(square)
```

**First, we create:**

1- (( openList )) form PriorityQueue class because we want to expand the node with the minimum evalution function (f). 


2- (( closedList )) insert expanded node to avoid repetion of already explored node.


3- object maze of type Maze and initialize it.


we took the start node from initialized maze and insert it in (( openList )) to begin search for the shortest path using A* algorithm concepts.


the while loop repeats until the openList is empty or the goal is found.
in each iteration square will updated with the node that has minimum value of evalution function.
then we check if the square node is end node ( goal ) so, we print the complete path and break out the loop.
if it's not the end node and in (( closedList )) we skip this iteration.


Otherwise, we get the adjecnts of this node and insert it to (( openList )).
finally, we insert this node to (( closedList )).


### part two Simulated Annealing:

**Local search:** the process of search that can often find reasonable solutions in large or infinite state spaces by keep track of single current state by moving only to the neighboring state . 
one powerfull point of local search is using little memory


**Complete vs incremental problem configurations:**

**complete problem:** predefined all states before we start.

**incremental problem:** generate states serially. 


**Simulated annealing:** is a searching algorithm solve Hill-climbing problem by avoiding stuck on a local maximum, allowing some "bad" moves but with some restrictions.

```python
T=1
maze = Maze()
maze.initMaze()          
alpha=0.9  

maze.filladj()
current=maze.start

#Simulated annealing is start !!!

#While the temperature keeps on decreasing, the Simulated Annealling algorithm will start making actions and moves to new states
while(T>0.0001):
    
    # try to change the range(20 in our example)
    for i in range(20):
        next=current.adjecnts[random.randint(0,(len(current.adjecnts))-1)]
        next.parent=copy.copy(current)
        
        #we find the goal
        if(next is maze.end):            
            maze.path(next)
            T=0   #to break the outer loop
            break
        E=next.h-current.h 
        
        # good move
        if(E<0):          
            current=copy.copy(next)
            
        #bad move 
        else:
            e= 2.72
            power= -E/T
            prob = pow(e,power)
            if(prob>random.random()):
                 current=copy.copy(next)   
    #decrease T       
    T=T*alpha
```

Now, we have “T” variable to control our iteration and the number allowed of bad moves.

Initially, the temperature"T" is set to a high value (1 in our example) and it keeps on decreasing until it is almost zero(0.0001 in our example).
 

Then, we created “maze” object from Maze class,defined “alpha” value, and created “current” variable and assigned start node of the maze to it. Then called **“filladj” method**.

We start our search here with tow loop, outer one controlled by ”T” variable.
Inner loop is used to prevent infinite loop, we created “next” node which is a random successor from ((adjecnts)) list of “current” node, and set its parent with the “current” node.
If the “next” node is the end of the maze we call **“path” method**,  make T = 0 to end the outer loop and break this loop.

Otherwise, calculate “E” which is the difference between next and current heuristic, if this E < 0 that’s means this move is a good one so we change “current” node to “next” node values. 


If E >= 0  that’s means this move is a bad one , we will calculate the probability "prob" which is decrease when the time is increasing, it's control the bad move depends on the value of T. After computing "prob" value we compare it with random number between [0,1)

if the "prob" is less than the random number we change “current” node to “next” node values. 

finally we decrease the "T" variable with alpha value.
